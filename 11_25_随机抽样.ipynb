创建分别具有5000个数据的训练集（xi，y）和测试集（xi，y），其中xi在间隔（-10,10）上均匀随机分布。为了使回归问题“真实”，大的均匀噪声分布在[-0.2,0.2]已添加到所有训练样本中，同时测试数据保持无噪声。

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.random.uniform(-10,10,5000)
y1=np.where(x!=0, np.sin(x)/x, 0)
noise=np.random.uniform(-0.2,0.2,size=5000)
y1=y1+noise

x2=np.random.uniform(-10,10,5000)
x2=np.sort(x2)
y2=np.where(x2!=0, np.sin(x2)/x2, 0)

plt.scatter(x, y1)
plt.plot(x2, y2, color='coral')
```
